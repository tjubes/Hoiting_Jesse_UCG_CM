```python
from math import exp
  
def f(c):
    return 668.06/c * (1-exp(-0.146843*c))-40

def getRoot(Xl, Xu, n):
    if f(Xl)*f(Xu)>=0:
        print("Inaccurate guess, function won't work")
        return None
    for i in range(n):
        root = (Xl+Xu)/2
        f_root = f(root)
        if f(Xl)*f_root < 0:
            Xl = Xl
            Xu = root
        elif f(Xu)*f_root < 0: 
            Xl = root
            Xu = Xu
        elif root == 0:
            print("Found exact solution")
            return root
        else:
            print("Bisection method fails")
            return None
    return (Xl+Xu)/2
```
